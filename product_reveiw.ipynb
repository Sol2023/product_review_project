```python
import os
import pandas as pd

import openai
from openai import OpenAI
```

```python


# Path to the folder containing CSV files
folder_path = "data/NPIC-review"

# Initialize an empty DataFrame to store combined data
combined_data = pd.DataFrame()

# Loop through each file in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        # Read the CSV file
        file_path = os.path.join(folder_path, file_name)
        data = pd.read_csv(file_path)
        
        # Extract product ID from file name
        product_id = os.path.splitext(file_name)[0]
        
        # Add product_id column
        data['product_id'] = product_id
        
        # Select and rename columns
        data = data.rename(columns={
            "a-profile-name": "user_name",
            "a-icon-alt": "rate_stars",
            "a-size-base": "review_title",
            "a-size-base 2": "review_date",
            "a-size-base 3": "comment_content"
        })[['product_id', 'user_name', 'rate_stars', 'review_title', 'review_date', 'comment_content']]
        
        # Append the data to the combined DataFrame
        combined_data = combined_data.append(data, ignore_index=True)

# Write the combined data to a new CSV file
combined_file_path = "data/review_data.csv"
combined_data.to_csv(combined_file_path, index=False)

print("Combined CSV file saved at:", combined_file_path)
```

## Data Cleaning

```python
# Transfer rate_stars to float type
df['rate_stars'] = df['rate_stars'].str.split(" ").str[0].astype(float)

# Fetch review datetime 
# Define regex pattern to match "Reviewed in [Country]"
pattern = r'Reviewed in (.*?) on'
# Extract country name and replace it with 'United States'
df['review_date'] = df['review_date'].str.replace(pattern, 'Reviewed in United States on')
# Convert the 'review_date' column to datetime
df['review_date'] = pd.to_datetime(df['review_date'], format='Reviewed in United States on %B %d, %Y')
```

## Sentiment Analysis

```python
df.info()
```

```python
from textblob import TextBlob

# Function to get sentiment polarity
def get_sentiment(comment):
    if isinstance(comment, str):
        analysis = TextBlob(comment)
        if analysis.sentiment.polarity > 0:
            return 'positive'
        elif analysis.sentiment.polarity == 0:
            return 'neutral'
        else:
            return 'negative'
    else:
        return 'neutral'

# Apply sentiment analysis and create a new column
df['sentiment'] = df['comment_content'].apply(get_sentiment)
```
